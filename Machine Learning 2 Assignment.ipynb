```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import sklearn
from sklearn.datasets import load_boston

#Load Data from the Dataset
boston = load_boston()

#Create a Dataframe consisting of all the feature, 13 in total
bos = pd.DataFrame(boston.data,columns=boston.feature_names)

# Put the Price in another DataFrame
price = pd.DataFrame(boston.target, columns=["MEDV"])
```

```python
bos.shape
```

```python
price.shape
```

```python
#import the linear Regression package
from sklearn.linear_model import LinearRegression

#Assign the independent Variables or the features. In this case we are dealing with Multiple linear regression
X = bos

#Assign the Dependent Variable.
Y = price

#We will fit the Linear model
lm = LinearRegression() 
lm.fit(X, Y)

# print intercept and coefficients 
print(lm.intercept_) 
print(lm.coef_)
```

```python
#Once the Linear model is fit, we will predict using the model and check the score
print(lm.predict(X))
```

```python
#R-Squared Score
lm.score(X,Y)
```
